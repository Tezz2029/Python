```python
# Examples of Arithmetic Operator
a = 9
b = 4

# Addition of numbers
add = a + b

# Subtraction of numbers
sub = a - b

# Multiplication of number
mul = a * b

# Division(float) of number
div1 = a / b

# Division(floor) of number
div2 = a // b

# Modulo of both number
mod = a % b

# Power
p = a ** b

# print results
print("Addition :- ", add)
print("Substraction :- ", sub)
print("Multiplication ", mul)
print("Division float :- ", div1)
print("Division floor :- ", div2)
print("Modulo :- ", mod)
print("Power :- ", p)
```

```python
# Examples of Relational Operators
a = 13
b = 33

# a > b is False
print("Is a Greater than b :- ", a > b)

# a < b is True
print("Is a Less than b :- ", a < b)

# a == b is False
print("Is a Equal's to b :- ", a == b)

# a != b is True
print("Is a Not Equal's to b :- ", a != b)

# a >= b is False
print("Is a Greater than or Equal to b :- ", a >= b)

# a <= b is True
print("Is a Less than or Equal to b :- ", a <= b)
```

```python
# Examples of Logical Operator
a = True
b = False

# Print a and b is False
print("a&b :- ", a and b)

# Print a or b is True
print("a|b :- ", a or b)

# Print not a is False
print("!a :-", not a)
```

```python
# Examples of Bitwise operators
a = 10
b = 4

# Print bitwise AND operation
print("Biwise AND :- ", a & b)

# Print bitwise OR operation
print("Bitwise OR :- ", a | b)

# Print bitwise NOT operation
print("Bitwise NOT :- ", ~a)

# print bitwise XOR operation
print("Bitwise XOR :- ", a ^ b)

# print bitwise right shift operation
print("Bitwise Right shift :- ", a >> 2)

# print bitwise left shift operation
print("Bitwise Left shift :- ", a << 2)
```

```python
# Examples of Assignment Operators
a = 10

# Assign value
b = a
print("Assign Value :- ", b)

# Add and assign value
b += a
print("Add and Assign value :- ", b)

# Subtract and assign value
b -= a
print("Subtract and Assign value :- ", b)

# multiply and assign
b *= a
print("Multiply and assign value :- ", b)

# bitwise lishift operator
b <<= a
print("Bitwise lestshift operator :- ", b)
```

```python
# Example of Identity Operators

a = 10
b = 20
c = a

print(a is not b)
print(a is c)
```

```python
# Example of Membership Operator
# Python program to illustrate
# not 'in' operator
x = 24
y = 20
list = [10, 20, 30, 40, 50]

print("not in Example")
if (x not in list):
	print("x is NOT present in given list")
else:
	print("x is present in given list")

print("******************************")

print("in Example")
if (y in list):
	print("y is present in given list")
else:
	print("y is NOT present in given list")
```

```python
# Examples of Operator Precedence

# Precedence of '+' & '*'
expr = 10 + 20 * 30
print(expr)

# Precedence of 'or' & 'and'
name = "Alex"
age = 0

if name == "Alex" or name == "John" and age >= 2:
	print("Hello! Welcome.")
else:
	print("Good Bye!!")
```

```python
# Examples of Operator Associativity

# Left-right associativity
# 100 / 10 * 10 is calculated as
# (100 / 10) * 10 and not
# as 100 / (10 * 10)
print(100 / 10 * 10)

# Left-right associativity
# 5 - 2 + 3 is calculated as
# (5 - 2) + 3 and not
# as 5 - (2 + 3)
print(5 - 2 + 3)

# left-right associativity
print(5 - (2 + 3))

# right-left associativity
# 2 ** 3 ** 2 is calculated as
# 2 ** (3 ** 2) and not
# as (2 ** 3) ** 2
print(2 ** 3 ** 2)
```
