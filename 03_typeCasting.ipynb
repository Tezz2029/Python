```python
# implicit type Casting

# Python automatically converts
# a to int
a = 7
print("Datatype of a is :- ", type(a))
print("*********************************")

# Python automatically converts
# b to float
b = 3.0
print("Datatype of b is :- ", type(b))
print("*********************************")

# Python automatically converts
# c to float as it is a float addition
c = a + b
print(c)
print("Datatype of c is :- ", type(c))
print("*********************************")

# Python automatically converts
# d to float as it is a float multiplication
d = a * b
print(d)
print("Datatype of d is :- ", type(d))
```

```python
# Python program to demonstrate Explicit type Casting
# int to float

# int variable
a = 5

# typecast to float
n = float(a)

print(n)
print("Datatype of n is :-", type(n))
```

```python
# Python program to demonstrate Explicit type Casting
# float to int

# int variable
a = 5.9

# typecast to int
n = int(a)

print(n)
print("Datatype of n is :-", type(n))
```

```python
# Python program to demonstrate Explicit type Casting
# int to str

# int variable
a = 5

# typecast to str
n = str(a)

print(n)
print("Datatype of n is :-", type(n))
```

```python
# Python program to demonstrate Explicit type Casting
# str to int

# string variable
a = "5"

# typecast to int
n = int(a)

print(n)
print("Datatype of n is :-", type(n))
```

```python
# Python program to demonstrate Explicit type Casting
# str to float

# string variable
a = "5.9"

# typecast to float
n = float(a)

print(n)
print("Datatype of n is :-", type(n))
```
