```python
# Strings in PYTHON
# textType Datatype
a = 'Tejas Lanjekar'
b = "I 'am Tejas' Lanjekar"
c = '''I "am Tejas" Lanjekar'''

print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

```python
#  Numeric type Datatypes
# int, float, complex

a = 8
b = 8.8
c = 9j


print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

```python
# Sequence type Datatypes
# list, tuple, range

a = ["apple", "banana", "cherry"]
b = ("apple", "banana", "cherry")
c = range(6)


print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

```python
# Mapping type Datatypes
# Dict

a = {"name" : "John", "age" : 36}

print(a)

print(type(a))
```

```python
# Set type Datatypes
# set & frozenset

a = {"apple", "banana", "cherry"}
b = frozenset({"apple", "banana", "cherry"})

print(a)
print(b)

print(type(a))
print(type(b))
```

```python
#  Boolean Datatype
# bool

a = True

print(a)

print(type(a))
```

```python
# Binary type Datatypes
# bytes, bytearray, memoryview

a = b"Hello"
b = bytearray(5)
c = memoryview(bytes(5))


print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

```python
# None Datatype

a = None

print(a)

print(type(a))
```
