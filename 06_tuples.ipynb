```python
# Creating Tuples
# An empty tuple
print("An empty tuple :-")
empty_tuple = ()
print (empty_tuple)
print("**********************************")

# Creating non-empty tuples
print("Creating non-empty tuples :-")
# One way of creation
tup = 'python', 'geeks'
print(tup)

# Another for doing the same
tup = ('python', 'geeks')
print(tup)
print("**********************************")

# Note: In case your generating a tuple with a single element, make sure to add a comma after the element.
```

```python
# Code for concatenating 2 tuples

tuple1 = (0, 1, 2, 3)
tuple2 = ('python', 'geek')

# Concatenating above two
print(tuple1 + tuple2)
print("**********************************")
```

```python
# Code for creating nested tuples

tuple1 = (0, 1, 2, 3)
tuple2 = ('python', 'geek')
tuple3 = (tuple1, tuple2)
print(tuple3)
```

```python
# Code to create a tuple with repetition

tuple3 = ('python',)*3
print(tuple3)
print("**********************************")

# Try the above without a comma and check. You will get tuple3 as a string ‘pythonpythonpython’.
tuple3 = ('python')*3
print(tuple3)
print("**********************************")
```

```python
#code to test that tuples are immutable

tuple1 = (0, 1, 2, 3)
tuple1[0] = 4
print(tuple1)
```

```python
# code to test slicing

tuple1 = (0 ,1, 2, 3)
print(tuple1[1:])
print("**********************************")

print(tuple1[::-1])
print("**********************************")

print(tuple1[2:4])
print("**********************************")
```

```python
# Code for deleting a tuple

tuple3 = ( 0, 1)
del tuple3
print(tuple3)
```

```python
# Code for printing the length of a tuple

tuple2 = ('python', 'geek')
print(len(tuple2))
```

```python
# Code for converting a list and a string into a tuple

list1 = [0, 1, 2]
print(tuple(list1))
print("**********************************")

print(tuple('python')) # string 'python'
print("**********************************")

# Takes a single parameter which may be a list,string,set or even a dictionary( only keys are taken as elements) and converts them to a tuple.
```

```python
#python code for creating tuples in a loop

tup = ('geek',)
n = 5 #Number of time loop runs
for i in range(int(n)):
	tup = (tup,)
	print(tup)
```

```python
# A python program to demonstrate the use of
# cmp(), max(), min()

tuple1 = ('python', 'geek')
tuple2 = ('coder', 1)

if (cmp(tuple1, tuple2) != 0):

	# cmp() returns 0 if matched, 1 when not tuple1
	# is longer and -1 when tuple1 is shorter
	print('Not the same')
else:
	print('Same')
print ('Maximum element in tuples 1,2: ' +
		str(max(tuple1)) + ',' +
		str(max(tuple2)))
print ('Minimum element in tuples 1,2: ' +
	str(min(tuple1)) + ',' + str(min(tuple2)))

```

```python
#  Count() Method
# Creating tuples
Tuple = (0, 1, (2, 3), (2, 3), 1,
		[3, 2], 'geeks', (0,))

# count the appearance of (2, 3)
res = Tuple.count((2, 3))
print('Count of (2, 3) in Tuple is:', res)
print("**********************************")

# count the appearance of [3, 2]
res = Tuple.count([3, 2])
print('Count of [3, 2] in Tuple is:', res)
print("**********************************")
```

```python
#Index() Method

# Example 1: Using Tuple Index() Method
# Creating tuples
Tuple = (0, 1, 2, 3, 2, 3, 1, 3, 2)

# getting the index of 3
res = Tuple.index(3)
print('First occurrence of 3 is', res)

# getting the index of 3 after 4th
# index
res = Tuple.index(3, 4)
print('First occurrence of 3 after 4th index is:', res)
print("**********************************")

# Example 2: Using Tuple() method when the element is not found
# Creating tuples
Tuple = (0, 1, 2, 3, 2, 3, 1, 3, 2)

# getting the index of 3
res = Tuple.index(4)

```
